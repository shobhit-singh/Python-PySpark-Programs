### pandas 03

In this notebook we will cover how we can convert Python dictionary to dataframe.

|S.No|Action|Command|Input|Output|
|:----|:-----|:-----:|:-----:|:-----:|
|1|Convert a Python Dict to Pandas Dataframe|pd.DataFrame(dict)|dict|pandas.core.frame.DataFrame (lets call this df)|
|2|Set the no of columns for print/head dataframe |pd.set_option('display.max_columns',n)|no of columns |None|
|3|Set the no of rows for print/head dataframe|pd.set_option('display.max_rows',n)|no of rows |None|
|4|Selecting the single column from dataframe|dataframe['name']/dataframe.name|column-name|pandas.core.series.Series Series is like 1-D array.  Series also have index.|
|5|Selecting the multiple column from dataframe|dataframe[[name1,name2...]]/dataframe.name|column-name list|pandas.core.frame.DataFrame|

```python
import pandas as pd
print("Let's start exploring", '\U0001F43C' , 'using',"\U0001F40D")
```

```python
import json
import requests

url = 'https://api.github.com/users/shobhit-singh/repos'
req = requests.get(url)
text = req.text

# create a dict using the json data
parsed_dict = json.loads(text)

# 1. Converting dict to pandas dataframe
df_git_repos = pd.DataFrame(parsed_dict)
```

```python
df_git_repos.head(2)
```

```python
df_git_repos.shape
```

9 rows & 73 columns. If you noticed above the output of df_git_repos.head(2) - it's not prining all the columns in jupyter notebook. Lets change the setting to print all the columns.

```python
# 2. Set the no of columns for print/head dataframe
pd.set_option('display.max_columns',73)
```

```python
# here you can see all the 73 columns
df_git_repos.head(2)
```

```python
# 4. Selecting the single column another way: df_git_repos['name']
df_git_repos.name
```

```python
type(df_git_repos.name)
```

Series is like 1-D array.  Series also have index.

```python
# 5. Selecting the multiple column from dataframe
df_git_repos[['name','description','url']]
```
